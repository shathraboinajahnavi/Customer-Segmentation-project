```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# ==========================================
# 1. GENERATE MOCK CUSTOMER DATA (For Testing)
# ==========================================
# If you have your own CSV file, replace this with: df = pd.read_csv('your_file.csv')
np.random.seed(42)

# Creating 3 distinct types of shoppers to simulate real behavior
group1 = np.random.normal(loc=[30, 20], scale=[5, 5], size=(30, 2))   # Low income, low spend
group2 = np.random.normal(loc=[60, 50], scale=[10, 10], size=(40, 2)) # Avg income, avg spend
group3 = np.random.normal(loc=[90, 80], scale=[7, 5], size=(30, 2))   # High income, high spend

features = np.vstack([group1, group2, group3])
df = pd.DataFrame(features, columns=["Annual_Income_k", "Spending_Score"])

print("--- Step 1: Customer Dataset Sample ---")
print(df.head(), "\n")

# ==========================================
# 2. DATA PREPROCESSING & SCALING
# ==========================================
# K-Means requires data to be on the same scale for accurate math
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)

# ==========================================
# 3. TRAIN K-MEANS CLUSTERING MODEL
# ==========================================
# We want to segment customers into 3 distinct groups
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
df["Cluster_Labels"] = kmeans.fit_predict(scaled_features)

print("--- Step 3: Segmentation Complete ---")
print(df.groupby("Cluster_Labels").mean(), "\n")

# ==========================================
# 4. VISUALIZE THE CUSTOMER SEGMENTS
# ==========================================
plt.figure(figsize=(10, 6))

colors = ["#FF5733", "#33FF57", "#3357FF"]
for i in range(num_clusters):
    cluster_data = df[df["Cluster_Labels"] == i]
    plt.scatter(
        cluster_data["Annual_Income_k"],
        cluster_data["Spending_Score"],
        s=100,
        c=colors[i],
        label=f"Segment {i}: " + ("Budget" if i==0 else "Average" if i==1 else "Premium Shoppers")
    )

# Plotting the cluster centers (centroids)
centers = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centers[:, 0], centers[:, 1], s=300, c="black", marker="*", label="Centroids")

plt.title("Customer Segmentation Analysis (K-Means)", fontsize=14)
plt.xlabel("Annual Income ($k)", fontsize=12)
plt.ylabel("Spending Score (1-100)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

# Save image locally in Colab environment
plt.savefig("customer_segments.png", dpi=300)
plt.show()
```
